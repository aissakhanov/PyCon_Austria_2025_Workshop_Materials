# Figure: quick practice

```python
# Create a Figure object
f = Figure()
f
```

```python
# add the data
trace = Scatter(x = [1,2,3], y = [1,2,3])
f.add_trace(trace)
f
```

```python
# add a bar chart
trace = Bar(x = [1,2,3], y = [1,2,3])
f.add_trace(trace)
f
```

```python
# change color of the 1st trace
f.data[0].marker.color = 'green'
f
```

```python
# change the data point in trace 2
modified_x = list(f.data[1].x)
modified_x[2] = -1
f.data[1].x = modified_x
f
```

```python
#add additional data point in trace 1
f.data[0].x += (5,)
f.data[0].y += (10,)
f
```

# Frames: quick practice

```python
f = Figure()
f.layout.updatemenus = [dict(
                        type="buttons",
                        buttons=[dict(label="Play",
                        method="animate",
                        args=[None])])]
trace = Scatter(x = [1,3], y = [1,3], mode = 'markers')
f.add_trace(trace)
f
```

```python
# add a trace with one point
trace = Scatter(x = [2], y = [2], mode = 'markers')
f.add_trace(trace)
f
```

```python
# create an animated line from trace 2 to 1st point of trace 1. Does it work?
frame_1_data = Scatter(x = [1,2], y = [1,2], mode = 'lines')
frame_1 = Frame(data = frame_1_data)
f.frames = [frame_1]
f
```

```python
# create a dummy inivisible trace that will be substituted by the animation
dummy_trace = Scatter(x = [0], y = [0], visible = False, showlegend = False)
f.add_trace(dummy_trace)
f
```

```python
# use dummy trace for the animation. Does it work?
frame_1_data = Scatter(x = [1,2], y = [1,2], mode = 'lines')
frame_1 = Frame(data = frame_1_data, traces = [2])
f.frames = [frame_1]
f
```

```python
# use dummy trace for the animation. Does it work?
frame_1_data = Scatter(x = [1,2], y = [1,2], mode = 'lines', visible = True, showlegend = True, name = 'distance line')
frame_1 = Frame(data = frame_1_data, traces = [2])
f.frames = [frame_1]
f
```

```python
# create animated lines to both points
frame_1_data = Scatter(x = [f.data[1].x[0],f.data[0].x[1]], y = [f.data[1].y[0], f.data[0].x[1]], mode = 'lines', visible = True, showlegend = True, name = 'distance line')
frame_1 = Frame(data = frame_1_data, traces = [2])
f.frames = f.frames + (frame_1,)
f
```
